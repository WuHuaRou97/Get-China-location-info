```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
mainUrl = 'http://www.stats.gov.cn/tjsj/tjbz/tjyqhdmhcxhfdm/2021/'
endUrl = '.html'
indexUrl = f'{mainUrl}index{endUrl}'
```

```python
# 输入 url
# 返回 url页面的表格
def f_urlTo_dirt(url):
    r = requests.get(url)
    r.encoding = 'utf-8'
    soup = BeautifulSoup(r.text)
    tag_a = soup.find_all(name = 'a')[:-1]
    d = {}
    for i in tag_a:
        d[i.text]=i['href']
    return d
```

```python
# 输入 字典，对字典进行筛选
# 返回 偶数的字典
def f_remove_int_in_dirt(d):
    d_filteried = d.copy()
    i = 0
    for ii in d:
        if i%2==0:
            del d_filteried[ii]    
        i+=1
    return d_filteried
```

```python
# 输入当前页面的 字典
# 返回扩展下一页的 字典
def f_nextPage(currentPage):
    for i in currentPage:
        if len(currentPage[i])<13:
            nextUrl = f'{mainUrl}{currentPage[i]}'
            nextd = f_urlTo_dirt(nextUrl)
            nextd_filteried = f_remove_int_in_dirt(nextd)

            if type(nextd_filteried) == type({}):
                f_nextPage(nextd_filteried)

            currentPage[i] = nextd_filteried

    return currentPage
```

```python
indexPage = f_urlTo_dirt(indexUrl)
indexPage   
```

```python
result= f_nextPage(indexPage)
```
